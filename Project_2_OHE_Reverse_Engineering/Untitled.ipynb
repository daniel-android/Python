```python
# One-hot encoding (OHE)
#OHE is a way to transform a categorical feature, which may consist of text labels in
#the raw data, into a numerical feature that can be used in mathematical models.

#Here we consider the opposite problem:
# Consider 1=graduate school
#          2=university
#          3=high school
#          4=others
#We would like to recreate a column that has these strings, instead
#of numbers

import pandas as pd
import matplotlib as mpl #additional plotting functionality
mpl.rcParams['figure.dpi'] = 400 #high resolution figures
data = pd.read_csv('Data.csv')
```

```python
data.head(10)
```

```python
#Create an empty column for the categorical labells
#called 'EDUCATION_CAT'
data['EDUCATION_CAT']='none'
```

```python
data.head(10)
```

```python
data[['EDUCATION','EDUCATION_CAT']].head(10)
```

```python
#We need to populate this new column with the appropriate strings.
#Creating a dictionary
cat_mapping ={1:'graduate school', 2: 'university', 3:'high school', 4:'others'}
#Creating a map between EDUCATION and EDUCATION_CAT
data['EDUCATION_CAT']=data['EDUCATION'].map(cat_mapping)
data[['EDUCATION','EDUCATION_CAT']].head(10)
```

#Now we are ready to one-hot encode

```python
#create a one-hot encoded DataFrame of the EDUCATION_CAT column
data['EDUCATION_CAT'].head(10)
```

```python
edu_df=pd.get_dummies(data['EDUCATION_CAT'])
edu_df.head(10)
#edu_df.shape
```

```python
#Let's concatenate the series EDUCATION_CAT with the 
#data frame edu_df (to see if it really worked)
edu_df_with_ohe=pd.concat([data,edu_df],axis=1)
edu_df_with_ohe[['EDUCATION_CAT','graduate school','high school', 'university','others']].head(10)
```

```python
edu_df_with_ohe.drop(['others','university','high school', 'graduate school'],axis=1).head(10)
```

```python
edu_df_with_ohe.to_csv('data_2.csv',index=False)
```
